# Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
scada_data = pd.read_csv('scada_data.csv')
fault_data = pd.read_csv('fault_data.csv')
status_data = pd.read_csv('status_data.csv')
```

```python
t = scada_data['Time']
max_wind_speed = scada_data['WEC: Production kWh']

plt.figure()
plt.plot(t, max_wind_speed)
```

```python
faults = ['AF', 'EF', 'FF', 'GF', 'MF']
count = [62, 174, 254, 43, 20]
bar_labels = ['Airbox Cooling', 'Electric', 'Feeding', 'Gearbox', 'Mains']
bar_colors = ['tab:blue', 'tab:purple', 'tab:red', 'tab:green', 'tab:orange']

fig, ax = plt.subplots()
ax.bar(faults, count, color=bar_colors, label=bar_labels)
ax.set_title("Number of Fault Classifications")
ax.set_xlabel('Fault Type')
ax.set_ylabel('Count')
plt.show()
```
